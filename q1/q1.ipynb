```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
points = pd.read_csv('ML_Q01.csv')
m_arr = [0]
c_arr = [0]

def cost_function(m0, m1, data):
    n = len(data)
    c = 0

    for i in range(n):
        x = data.iloc[i].x
        y = data.iloc[i].y
        c += ((m1 * x) + m0 - y) ** 2

    c = c / (2 * n)
    return c
```

```python
def gradient_descent(m0, m1, data):
    m1_gradient = 0
    m0_gradient = 0
    n = len(data)

    for i in range(n):
        x = data.iloc[i].x
        y = data.iloc[i].y

        m0_gradient += -((m1 * x) + m0 - y)
        m1_gradient += -((m1 * x) + m0 - y) * x

    m0 = m0 + m0_gradient * (L / n)
    m1 = m1 + m1_gradient * (L / n)
    return m0, m1
```

```python
m1 = 1
m0 = 1
L  = 0.01
c_arr = []
iterations = 200

for j in range(iterations):
    #m_arr.append(m1)
    c = cost_function(m0, m1, points)
    c_arr.append(c)
    m0, m1 = gradient_descent(m0, m1, points)
    print("m{}, b{},cost{}, iterations{}" .format(m1, m0,c,j))
    plt.plot(list(range(0, 10)), [(m1 * x) + m0 for x in range(0, 10)], color="red")

plt.scatter(points.x, points.y, color = "pink")

plt.show()
```

```python
x_axis = np.linspace(0,200, 200)
plt.plot(x_axis,c_arr)
```

```python

df = pd.read_csv('ML_Q01.csv')
x = df['x'].values
y = df['y'].values
x = x.reshape(-1, 1)

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
x_poly = poly_features.fit_transform(x)

reg = LinearRegression()
reg.fit(x_poly, y)

x_vals = np.linspace(-10, 10, 20).reshape(-1, 1)
x_vals_poly = poly_features.transform(x_vals)

y_vals = reg.predict(x_vals_poly)

plt.scatter(x, y)
plt.plot(x_vals, y_vals, color = "red")
plt.show()
```
